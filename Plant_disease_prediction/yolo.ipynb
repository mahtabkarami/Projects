```python

# ============================
# 1. Install Dependencies
# ============================
!pip install roboflow ultralytics --quiet

# ============================
# 2. Download Dataset from Roboflow
# ============================
from roboflow import Roboflow

# Replace with your actual API key from https://roboflow.com
rf = Roboflow(api_key="EkfsugoPz76hituogORj")  # <-- Replace this

# Load the dataset
project = rf.workspace("dagi").project("crop-monitoring")
dataset = project.version(4).download("yolov8")

# Optional: check dataset path
import os
dataset_path = dataset.location
print("Dataset downloaded to:", dataset_path)

# ============================
# 3. Train YOLOv8
# ============================
from ultralytics import YOLO

# Load a pretrained YOLOv8 model (Nano version for speed; use 'yolov8s.pt' for better accuracy)
model = YOLO("yolov8n.pt")

# Train the model
results = model.train(
    data=os.path.join(dataset_path, "data.yaml"),
    epochs=50,
    imgsz=640,
    batch=16
)

# ============================
# 4. Evaluate Model (Optional)
# ============================
metrics = model.val()

# ============================
# 5. Run Prediction on Image
# ============================
from IPython.display import Image

# Upload an image or use an example from your dataset
image_path = os.path.join(dataset_path, "/Internal storage/Download/Crop monitoring.v4-raw-images.yolov8/valid/images/20200610_064003_jpg.rf.74ccbf9ecb8f332472c35949bd5bd7ce.jpg")# Path to test images

# List available test images
import glob
test_images = glob.glob(image_path + "/*.jpg")
print("Sample test image:", test_images[0])

# Predict and show
results = model.predict(source=test_images[0], save=True)

# Display result image
output_path = results[0].save_dir + "/" + os.path.basename(test_images[0])
Image(filename=output_path)
```
